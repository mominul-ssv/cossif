## Filter Images

```python
t_akiec = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_akiec.npy'), allow_pickle=True)
t_bcc = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_bcc.npy'), allow_pickle=True)
t_bkl = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_bkl.npy'), allow_pickle=True)
t_df = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_df.npy'), allow_pickle=True)
t_mel = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_mel.npy'), allow_pickle=True)
t_vasc = np.load(os.path.join(save_path, 'HAM10000_FBGT_t_vasc.npy'), allow_pickle=True)

loaded_records = [
    {'class': 'akiec', 'records': t_akiec, 'type': 'similar'}, 
    {'class': 'bcc', 'records': t_bcc, 'type': 'similar'}, 
    {'class': 'bkl', 'records': t_bkl, 'type': 'similar'}, 
    {'class': 'df', 'records': t_df, 'type': 'similar'}, 
    {'class': 'mel', 'records': t_mel, 'type': 'similar'}, 
    {'class': 'vasc', 'records': t_vasc, 'type': 'similar'}, 
]

# α=0.80 [alpha_1]
# α=0.85 [alpha_2]
# α=0.90 [alpha_3]

hp_alpha = [
    {'id':'alpha_1', 'value': 0.80}, 
    {'id':'alpha_2', 'value': 0.85}, 
    {'id':'alpha_3', 'value': 0.90}
]

for i in loaded_records:
    for alpha in hp_alpha:
        print(f"{i['class']} || {alpha['id']} || {'{:.2f}'.format(alpha['value'])}")
        source_path = os.path.join(root, f"split/train/{i['class']}")
        filtered_path = os.path.join(root, f"stylegan2-ada/oversampling/FBGT/{alpha['id']}/filtered/{i['class']}")
        removed_path = os.path.join(root, f"stylegan2-ada/oversampling/FBGT/{alpha['id']}/removed/{i['class']}")

        cossif.export_filtered_images(
            source_dir=source_path, 
            filtered_dir=filtered_path, 
            removed_dir=removed_path, 
            set_R=i['records'], 
            remove=i['type'],
            alpha=alpha['value']
        )
```
