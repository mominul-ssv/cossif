```python
####################################
# target class: akiec              #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: akiec           #
# type: real images                #
####################################
cossif.calculate_similarities(
    t_path=target_cls_akiec_alpha_1, 
    s_path=secondary_cls_akiec, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_akiec_alpha_1', 
    image_size=64,
    record_range=1
)

####################################
# target class: bcc                #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: bcc             #
# type: real images                #
####################################

cossif.calculate_similarities(
    t_path=target_cls_bcc_alpha_1, 
    s_path=secondary_cls_bcc, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_bcc_alpha_1', 
    image_size=64,
    record_range=1
)

####################################
# target class: bkl                #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: bkl             #
# type: real images                #
####################################
cossif.calculate_similarities(
    t_path=target_cls_bkl_alpha_1, 
    s_path=secondary_cls_bkl, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_bkl_alpha_1', 
    image_size=64,
    record_range=1
)

####################################
# target class: df                 #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: df              #
# type: real images                #
####################################
cossif.calculate_similarities(
    t_path=target_cls_df_alpha_1, 
    s_path=secondary_cls_df, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_df_alpha_1', 
    image_size=64,
    record_range=1
)

####################################
# target class: mel                #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: mel             #
# type: real images                #
####################################
cossif.calculate_similarities(
    t_path=target_cls_mel_alpha_1, 
    s_path=secondary_cls_mel, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_mel_alpha_1', 
    image_size=64,
    record_range=1
)

####################################
# target class: vasc               #
# type: synthetic images           #
# distribution: α = 0.75 | alpha_1 #
####################################
# secondary class: vasc            #
# type: real images                #
####################################
cossif.calculate_similarities(
    t_path=target_cls_vasc_alpha_1, 
    s_path=secondary_cls_vasc, 
    save_path=save_path, 
    file_name='HAM10000_FAGT_t_vasc_alpha_1', 
    image_size=64,
    record_range=1
)
```

```python
t_akiec = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_akiec_alpha_1.npy'), allow_pickle=True)
t_bcc = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_bcc_alpha_1.npy'), allow_pickle=True)
t_bkl = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_bkl_alpha_1.npy'), allow_pickle=True)
t_df = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_df_alpha_1.npy'), allow_pickle=True)
t_mel = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_mel_alpha_1.npy'), allow_pickle=True)
t_vasc = np.load(os.path.join(save_path, 'HAM10000_FAGT_t_vasc_alpha_1.npy'), allow_pickle=True)

loaded_records = [
    {'class': 'akiec', 'records': t_akiec, 'type': 'similar'}, 
    {'class': 'bcc', 'records': t_bcc, 'type': 'similar'}, 
    {'class': 'bkl', 'records': t_bkl, 'type': 'similar'}, 
    {'class': 'df', 'records': t_df, 'type': 'similar'}, 
    {'class': 'mel', 'records': t_mel, 'type': 'similar'}, 
    {'class': 'vasc', 'records': t_vasc, 'type': 'similar'}, 
]

alpha = {'id':'alpha_1', 'value': 0.75}

for i in loaded_records:
    print(f"{i['class']} || {alpha['id']} || {'{:.2f}'.format(alpha['value'])}")
    source_path = os.path.join(root, f"stylegan2-ada/oversampling/FAGT/{alpha['id']}/oversample/{i['class']}")
    filtered_path = os.path.join(root, f"stylegan2-ada/oversampling/FAGT/{alpha['id']}/filtered/{i['class']}")
    removed_path = os.path.join(root, f"stylegan2-ada/oversampling/FAGT/{alpha['id']}/removed/{i['class']}")

    cossif.export_filtered_images(
        source_dir=source_path, 
        filtered_dir=filtered_path, 
        removed_dir=removed_path, 
        set_R=i['records'], 
        remove=i['type'],
        alpha=alpha['value']
    )
```
